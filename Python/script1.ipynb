```python
import pandas as pd
# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_seq_items', None)
# pd.set_option('display.width', None)
```

```python
df_step1_1=pd.read_csv("data/step1.1.csv")
```

```python
df_step1_1.columns
```

```python
df_covariates = pd.DataFrame()
df_covariates[['gvkey','datadate','fyear','cusip','ipodate', 'conm', 'at','sale','ebit','re','reuna','wcap','mkvalt','lt']] = df_step1_1[['gvkey','datadate','fyear','cusip', 'ipodate','conm', 'at','sale','ebit','re','reuna','wcap','mkvalt','lt']]   
df_covariates["liquidity"] = df_step1_1['wcap']/df_step1_1['at']
df_covariates["re/at"] = df_step1_1['re']/df_step1_1['at']
df_covariates["reuna/at"] = df_step1_1['reuna']/df_step1_1['at']
df_covariates["profitability"] = df_step1_1['ebit']/df_step1_1['at']
df_covariates["mkvalt/lt"] = df_step1_1['mkvalt']/df_step1_1['lt']
df_covariates["AssetTurnoverRatio"] = df_step1_1['sale']/df_step1_1['at']
df_covariates["issuer"]=df_covariates["cusip"].str[:6]
df_covariates["issue"]=df_covariates["cusip"].str[6:8]
df_covariates["issuer+issue"]=df_covariates["cusip"].str[:8]
df_covariates['check_digit'] = df_covariates['cusip'].str[8:9]
df_covariates
```

```python
print(df_covariates["issuer"].nunique())
print(df_covariates["issuer+issue"].nunique())
```

```python
df_step2=pd.read_csv("data/step2.csv")
df_step2_copy = df_step2.copy()
df_step2_copy
```

```python
df_step2_copy["issuer+issue"]=df_step2_copy["CUSIP"].str[:8]
print(df_step2_copy["issuer+issue"].nunique())
```

To Merge the Two data sets

```python
df_final_data = pd.merge(df_covariates, df_step2_copy, on='issuer+issue', how='inner')
```

# Trying to play with the data and trying to find survival time

```python
df_final_data_copy = df_final_data.copy()
df_final_data_copy.drop(columns=['reuna','issuer','issue','check_digit','issuer+issue','gvkey'], inplace=True)
df_final_data_copy['datadate']= pd.to_datetime(df_final_data_copy['datadate'])
df_final_data_copy['fyear'] =  df_final_data_copy.apply(lambda row: row['fyear'] if not pd.isna(row['fyear']) else row['datadate'].year,axis=1)
df_final_data_copy['fyear'] = df_final_data_copy['fyear'].astype(int)
```

```python
df_final_data_copy_to_impute.replace([np.inf, -np.inf], np.nan, inplace=True)
df_final_data_copy_to_impute.fillna(0, inplace=True)
```

```python
df_final_data_copy_to_impute.describe()
```

```python

clf = SoftImpute(J=2, lambda_=0.0)
fit = clf.fit(df_final_data_copy_to_impute.values)
df_final_data_copy_impute = clf.predict(df_final_data_copy_to_impute.values)
df_final_data_copy_impute
```

```python
df_final_data_copy[['liquidity','re/at','reuna/at','profitability','mkvalt/lt','AssetTurnoverRatio']] = pd.DataFrame(df_final_data_copy_impute, columns=['liquidity','re/at','reuna/at','profitability','mkvalt/lt','AssetTurnoverRatio'])
```

```python
selected_columns = df_pivot1.iloc[:, 9:]
# Count the number of NaN values in each row
nan_counts_per_row = selected_columns.isna().sum(axis=1)
print("Number of NaN values in each row:")
print(nan_counts_per_row)
```

```python
df = pd.DataFrame(selected_columns)
df[(nan_counts_per_row>180) & (nan_counts_per_row<300)]
print(df.shape[0])
print(df[(nan_counts_per_row>220)].shape[0])
print(df[(nan_counts_per_row>210)].shape[0])
print(df[(nan_counts_per_row>200)].shape[0])
print(df[(nan_counts_per_row>190)].shape[0])
print(df[(nan_counts_per_row>180)].shape[0])
print(df[(nan_counts_per_row>170)].shape[0])
print(df[(nan_counts_per_row>160)].shape[0])
print(df[(nan_counts_per_row>150)].shape[0])
print(df[(nan_counts_per_row>140)].shape[0])
print(df[(nan_counts_per_row>130)].shape[0])
print(df[(nan_counts_per_row>120)].shape[0])
```

```python
# Identify the row indices where the count of NaN values is greater than 190
rows_to_remove = nan_counts_per_row[nan_counts_per_row > 190].index
df_pivot_after_dropping_nans= df_pivot1.drop(rows_to_remove)
df_pivot_after_dropping_nans
```

Repeating imputing process on this

```python
import numpy as np
from py_soft_impute.soft_impute import SoftImpute
df_pivot_after_dropping_nans_to_impute = df_pivot_after_dropping_nans.iloc[:, 9:]
df_pivot_after_dropping_nans_to_impute.replace([np.inf, -np.inf], np.nan, inplace=True)
df_pivot_after_dropping_nans_to_impute.fillna(0, inplace=True)
clf = SoftImpute(J=2, lambda_=0.0)
fit = clf.fit(df_pivot_after_dropping_nans_to_impute.values)
df_pivot_after_dropping_nans_imputed = clf.predict(df_pivot_after_dropping_nans_to_impute.values)
print(df_pivot_after_dropping_nans_imputed.shape)
new_df1 = pd.DataFrame(df_pivot_after_dropping_nans_imputed, columns=df_pivot_after_dropping_nans_to_impute.columns)

df_pivot_after_dropping_nans_imputed= df_pivot_after_dropping_nans.copy()
df_pivot_after_dropping_nans_imputed.iloc[:, 9:] = new_df1
```
