```python
import pandas as pd
# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_seq_items', None)
# pd.set_option('display.width', None)
```

```python
df_step1_1=pd.read_csv("data/step1.1.csv")
```

```python
df_step1_1.columns
```

```python
df_covariates = pd.DataFrame()
df_covariates[['gvkey','datadate','fyear','cusip','ipodate', 'conm', 'at','sale','ebit','re','reuna','wcap','mkvalt','lt']] = df_step1_1[['gvkey','datadate','fyear','cusip', 'ipodate','conm', 'at','sale','ebit','re','reuna','wcap','mkvalt','lt']]   
df_covariates["liquidity"] = df_step1_1['wcap']/df_step1_1['at']
df_covariates["re/at"] = df_step1_1['re']/df_step1_1['at']
df_covariates["reuna/at"] = df_step1_1['reuna']/df_step1_1['at']
df_covariates["profitability"] = df_step1_1['ebit']/df_step1_1['at']
df_covariates["mkvalt/lt"] = df_step1_1['mkvalt']/df_step1_1['lt']
df_covariates["AssetTurnoverRatio"] = df_step1_1['sale']/df_step1_1['at']
df_covariates["issuer"]=df_covariates["cusip"].str[:6]
df_covariates["issue"]=df_covariates["cusip"].str[6:8]
df_covariates["issuer+issue"]=df_covariates["cusip"].str[:8]
df_covariates['check_digit'] = df_covariates['cusip'].str[8:9]
df_covariates
```

```python
print(df_covariates["issuer"].nunique())
print(df_covariates["issuer+issue"].nunique())
```

```python
df_step2=pd.read_csv("data/step2.csv")
df_step2_copy = df_step2.copy()
df_step2_copy
```

```python
df_step2_copy["issuer+issue"]=df_step2_copy["CUSIP"].str[:8]
print(df_step2_copy["issuer+issue"].nunique())
```

To Merge the Two data sets

```python
df_final_data = pd.merge(df_covariates, df_step2_copy, on='issuer+issue', how='inner')
```

```python
df_final_data
```

```python
df_final_data["DLSTCD"].nunique()
```

Trying to play with the data and trying to find survival time

```python
df_final_data_copy = df_final_data.copy()
df_final_data_copy.drop(columns=['reuna','issuer','issue','check_digit','issuer+issue','gvkey'], inplace=True)
df_final_data_copy['datadate']= pd.to_datetime(df_final_data_copy['datadate'])
df_final_data_copy['fyear'] =  df_final_data_copy.apply(lambda row: row['fyear'] if not pd.isna(row['fyear']) else row['datadate'].year,axis=1)
df_final_data_copy['fyear'] = df_final_data_copy['fyear'].astype(int)
```

```python
df_final_data_copy
```

```python
# Exporting dataframe as csv
df_final_data_copy.to_csv("data/data_with_ratios.csv", index=False)
```

```python
df_final_data_copy = df_final_data.copy()
df_final_data_copy.drop(columns=['CUSIP', 'at','sale','ebit','re','reuna','wcap','mkvalt','lt','issuer','issue','check_digit','issuer+issue','gvkey'], inplace=True)
df_final_data_copy['datadate']= pd.to_datetime(df_final_data_copy['datadate'])
df_final_data_copy['fyear'] =  df_final_data_copy.apply(lambda row: row['fyear'] if not pd.isna(row['fyear']) else row['datadate'].year,axis=1)
df_final_data_copy['fyear'] = df_final_data_copy['fyear'].astype(int)
df_final_data_copy
```

```python
df_pivot = df_final_data_copy.pivot_table(index=['cusip','conm','PERMNO','DLSTCD','HCOMNAM','HEXCD','BEGDAT','ENDDAT','ipodate'], columns='fyear', values=['liquidity','re/at','reuna/at','profitability','mkvalt/lt','AssetTurnoverRatio'])
df_pivot.columns= [f'{year}-{ratio}' for ratio, year in df_pivot.columns]
# Reset the index to make 'Company' a column again
df_pivot.reset_index(inplace=True)

# Display the reshaped DataFrame
df_pivot
```

```python
# Exporting dataframe as csv
df_pivot.to_csv("data/data_with_all_columns.csv", index=False)
```
